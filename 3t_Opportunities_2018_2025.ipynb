```python
import pandas as pd
import numpy as np


from google.colab import drive
drive.mount("/content/drive")

# Change directory to your specific folder
%cd /content/drive/MyDrive/3t

# Read the Excel file
data = pd.read_excel("3t_All_Opportunities_2018-2025.xlsx")

# Display the first few rows
data.head()
```

```python
data.info()
```

```python
duplicates = data[data.duplicated()]
print(duplicates)
```

```python
data.drop_duplicates()
```

```python
data['Opportunity Number'] = pd.to_numeric(data['Opportunity Number'], errors='coerce')
data.info()
```

```python
data['3t Division'] = data['3t Division'].replace("3t - Digital", "3t Digital")
data['3t Division'] = data['3t Division'].replace("3t - EnerMech", "3t EnerMech")
data['3t Division'] = data['3t Division'].replace("3t - GTSC", "3t GTSC")
data['3t Division'] = data['3t Division'].replace("3t - ALL STOP!", "3t ALL STOP")

data['Opportunity Sector'] = data['Opportunity Sector'].replace("Oil and Gas – Downstream", "Oil and Gas - Downstream")
data['3t Department']= data['3t Division'].str.split("-").str[0]
data['3t Department Type']= data['3t Division'].str.split("-").str[1]
data['Sector']= data['Opportunity Sector'].str.split("-").str[0]
data['Sector Type']= data['Opportunity Sector'].str.split("-").str[1]

data['3t Department Type'].unique()
```

```python
data_clean= data[~((data["3t Division"].isna()) & (data["Opportunity Sector"].isna()) & (data["Status"] == "Lost"))]

data_clean.info()
```

```python
data_clean['Opportunity Sector'].fillna('Unknown', inplace=True)
data_clean['Country'].fillna('Unknown', inplace=True)
data_clean['Client'].fillna('Unknown', inplace=True)
data_clean['3t Division'].fillna('Unknown', inplace=True)

data_clean.isna().sum()

```

```python
# data_clean.to_excel('data_clean.xlsx', index=False)
```

```python
# 1. Total number of columns
total_columns = data_clean.shape[1]

# 2. Number of numeric columns and their names
numeric_columns = data_clean.select_dtypes(include=['number']).columns

# 3. Number of categorical columns and their names
categorical_columns = data_clean.select_dtypes(include=['object']).columns

# 4. Number of datetime columns and their names
datetime_columns = data_clean.select_dtypes(include=['datetime']).columns

# 5. Number of rows
total_rows = data_clean.shape[0]

#6.
division= data_clean['3t Division'].unique()

#7.
sector= data_clean['Opportunity Sector'].unique()

print(f"Total number of columns: {total_columns}")
print(f"Total number of rows: {total_rows}")
print(f"Number of numeric columns: {len(numeric_columns)}, Names: {numeric_columns.tolist()}")
print(f"Number of categorical columns: {len(categorical_columns)}, Names: {categorical_columns.tolist()}")
print(f"Number of datetime columns: {len(datetime_columns)}, Names: {datetime_columns.tolist()}")

print(f"3t Divisions: {division}")
print(f"Opportunity Sectors: {sector}")
```
